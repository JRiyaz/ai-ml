```python
import numpy as np
import matplotlib.pyplot as plt

def plot_boundary_and_w(w, b):
    # Create a grid of points (x1) to plot the line against
    x_range = np.linspace(-5, 5, 100)
    
    # Calculate the boundary line: w1*x1 + w2*x2 + b = 0
    # Rearranging for x2 (y-axis): x2 = -(w1*x1 + b) / w2
    y_range = -(w[0] * x_range + b) / w[1]
    
    plt.figure(figsize=(8, 8))
    
    # 1. Plot the Decision Boundary
    plt.plot(x_range, y_range, 'k-', linewidth=2, label='Decision Boundary ($w^Tx + b = 0$)')
    
    # 2. Plot the Weight Vector (w)
    # We start the arrow from a point on the line to see the perpendicularity
    start_point = np.array([0, -b/w[1]]) 
    plt.quiver(start_point[0], start_point[1], w[0], w[1], angles='xy', scale_units='xy', scale=1, color='r', label='Weight Vector ($w$)')
    
    # 3. Visualize Half Spaces
    # Fill Positive Half Space (Where w points) - Cyan
    plt.fill_between(x_range, y_range, 6, color='cyan', alpha=0.3, label='Positive Half Space ($>0$)')
    # Fill Negative Half Space (Opposite to w) - Pink
    plt.fill_between(x_range, y_range, -6, color='magenta', alpha=0.1, label='Negative Half Space ($<0$)')

    # Graph Formatting
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid(True, linestyle='--')
    plt.legend(loc='lower right')
    plt.title(f"Geometry of ML: w={w}, b={b}")
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.show()

# --- Define Logic ---
# w = [1, 2] means for every 1 step right, we go 2 steps up.
# This vector points into the "Positive" region.
w_vec = np.array([1, 2]) 
bias = -2

plot_boundary_and_w(w_vec, bias)
```
